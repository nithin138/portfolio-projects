```python
#THE BETTER WAY TO UNDERSTAND INFORMATION IS THROUGH VISUALS
#PLOTTING GRAPHS BETWEEN DIFFERENT ATTRIBUTER HELPS TO IDENTIFY THE KEY PERFORMANCE INDICATORS

#NOW WE TRY TO PLOT A 'DISTRIBUTION PLOT' TO VISUALIZE THE DISTRIBUTION OF DATA
plt.figure(figsize=(10,4))
n=0
for x in ['Age','Annual Income (k$)','Spending Score (1-100)']:
    n+=1
    plt.subplot(1, 3, n)
    plt.subplots_adjust(hspace=0.5,wspace=0.5)
    sns.distplot(df[x],bins = 20)
    plt.title('distplot of {}'.format(x))
plt.show()
```

```python
#VALUE COUNTING THE POPULATION BASED ON GENDER

plt.figure(figsize=(8,2))
sns.countplot(y = 'Gender',data = df,palette='bright')
plt.show()
```

```python
#VIOLIN PLOT COMES INTO PLAY TO UNDERSTAND THE DISTRIBUTION OF DATA BY A ATTRIBUTE BASED ON THE OTHER ONE
#IN THIS PLOT WE IDENTIFY THE DISTRIBUTION OF PEOPLE IN DIFFERENT GENDERS BASED ON THEIR ANNUAL INCOME
plt.figure(figsize=(11,5))
n=0
for cols in ['Age','Annual Income (k$)','Spending Score (1-100)']:
    n+=1
    plt.subplot(1, 3, n)
    sns.set_style(style='white')
    plt.subplots_adjust(hspace=0.5,wspace=0.5)
    sns.violinplot(x =cols,hue = 'Gender',data = df,palette="bright")
    plt.ylabel('Gender' if n ==1 else '')
    plt.title('violinplot')
plt.show()
```

```python
#RELATIONSHIP BETWEEN SPENDING SCORE AND ANNUAL INCOMOE
#WE SEE THAT PEOPLE WITH ANNUAL INCOME BETWEEN 40 - 60 THOUSAND DOLLARD HAVE MOSTLY A SPENDING CORE BETWEEN 40 AND 60
sns.relplot(x = 'Annual Income (k$)', y = 'Spending Score (1-100)',data =df)
```

K - MEANS CLUSTERING

```python
#DONE WITH ANALYSIS WE NOW MOVE TOWARDS CLUSTERING THE DATA USING K MEANS CLUSTERING ALGORITHM

#HERE WE TRY TO BUILD CLUSTERS BASED ON DIFFERENT ASPECTS LIKE AGE,INCOME AND SPENDING SCORE
x1 = df.loc[:,["Age","Spending Score (1-100)"]].values

from sklearn.cluster import KMeans
wcss = []
for k in range(1,11):
    Kmeans =KMeans(n_clusters=k,init='k-means++') 
    Kmeans.fit(x1)
    wcss.append(Kmeans.inertia_)
plt.figure(figsize = (8,4))
plt.grid()
plt.plot(range(1,11),wcss,linewidth = 2,color="red",marker = "8")
plt.xlabel("k value")
plt.ylabel("wcss")
plt.show()
```

```python
#WE SEE THAT THERE IS SHRINK IN THE ABOVE GRAPH AT THE POINT 4 BY WHICH WE CAN ASSUME THAT IT IS BETTER TO BUILD 4 CLUSTERS FROM THE DATA
kmeans = KMeans(n_clusters=4)
label = kmeans.fit_predict(x1)#FITTING DATA INTO CLUSTERS BASING THE LABELS
print(label)
```

```python
print(kmeans.cluster_centers_)#PRINTING THE CLUSTER CENTROIDS
```

```python
plt.scatter(x1[:,0],x1[:,1],c = kmeans.labels_,cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color ='black')
plt.title('cluster of customers')
plt.xlabel("Age")
plt.ylabel("Spending scores")
plt.show()
#RREPRESENTING THE DATAPOINTS AS A SCATTER PLOT WITH DIFFERENTIATIVE COLORS REPRESENTING DIFFERENT CLUSTERS
#THE BLACK DOT IN BETWEEN IS THE CLUSTER CENTROID
```
