```python
df = df[['id','title','overview','genre']]
```

```python
df
```

```python
df['tags'] = df['overview']+df['genre']
```

```python
df
```

```python
df1 = df.drop(['overview','genre'],axis =1)
```

```python
df
```

```python
df1
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=10000,stop_words='english')
```

```python
cv
```

```python
vector = cv.fit_transform(df1['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
df1[df1['title']=="The Godfather"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])),reverse=True,key = lambda vector:vector[1])
for i in distance[0:5]:
    print(df1.iloc[i[0]].title)
```

```python
def recomend(df):
    index = df1[df1['title']==df].index[0]
    distance = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda vector:vector[1])
    for i in distance[0:5]:
        print(df1.iloc[i[0]].title)
```

```python
recomend("Iron Man")
```
