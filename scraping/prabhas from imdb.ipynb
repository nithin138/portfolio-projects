```python
from bs4 import  BeautifulSoup
import requests
```

```python
url = 'https://www.imdb.com/list/ls085709154/'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html')
page
```

```python
obj = soup.find('div',class_ = "lister-list") 
scrapped_titles = obj.find_all('h3') 



old_titles =[title.get_text().strip().replace('\n','') for title in scrapped_titles] 
first_sagam = old_titles[0:9]
tharwaatha_sagam = old_titles[9:] 
tiles1 = [item.lstrip(item[:2]) for item in first_sagam]
tiles2 = [item.lstrip(item[:3]) for item in tharwaatha_sagam]
titles = tiles1+tiles2
```

```python
release_year =[item[-5:-1] for item in titles]
print(release_year)
movie_name = [item[:-6]  for item in titles]
print(movie_name)
```

```python
scrapped_ratings = obj.find_all('span',class_ = 'ipl-rating-star__rating') ; print(type(scrapped_ratings))
old_ratings =[rating.get_text().strip().replace('\n','') for rating in scrapped_ratings] 
new_ratings = old_ratings[::23] ;new_ratings
```

```python
scrapped_genres = obj.find_all('span' , class_ = 'genre') 
genres = [genre.get_text().strip() for genre in scrapped_genres] ; genres
```

```python
scrapped_runtimes = obj.find_all('span' , class_ = 'runtime') 

runtime = [item.get_text() for item in scrapped_runtimes] 
runtime.insert(0,'162 min')
```

```python
scrapped_directors = obj.find_all('p',class_ = 'text-muted text-small') ; type(scrapped_directors)
old_directors = [director.get_text().strip().replace('\n','') for director in scrapped_directors] 
new_directors = old_directors[1::3]
new_directors = [x.split('    ',1) for x in new_directors]
director = []
i=0
while i<len(new_directors):
    director.append(new_directors[i][0])
    i+=1
director = [item[9:] for item in director]
director = [item[:-1] for item in director]
director = [item[0:] if item[0]!=':' else item[1:] for item in director ]
director
```

```python
cast = []
i=0
while i<len(new_directors):
    cast.append(new_directors[i][1])
    i+=1
cast = [item[7:] for item in cast] 
```

```python

votes = old_directors[2::3] ; len(new_directors);votes




votes = [item_votes[6:] if len(item_votes)<12 else item_votes[7:14] for item_votes in votes];votes
```

```python
scrapped_verdicts_and_shares = obj.find_all('div' ,class_ = 'list-description') 

old_verdicts_and_shares = [item.get_text().strip() for item in scrapped_verdicts_and_shares]




market_share = [item.split('V',1) for item in old_verdicts_and_shares]
i=0
scrapped_shares = []
while i<len(market_share):
    scrapped_shares.append(market_share[i][0])
    i+= 1
share = [item[25:] for item in scrapped_shares ]



verdicts = [item.split('r',3) for item in old_verdicts_and_shares]
i=0
scrapped_verdicts = []
while i<len(verdicts):
    scrapped_verdicts.append(verdicts[i][3])
    i+=1
verdict = [item[10:] for item in scrapped_verdicts]
```

```python
print(len(movie_name))
print(len(release_year))
print(len(director))
print(len(new_ratings))
print(len(runtime))
print(len(genres))
print(len(cast))
print(len(votes))
print(len(share))
print(len(verdict))
```

```python
import pandas as pd
df = pd.DataFrame({'moviename':movie_name,'release year':release_year,'rating':new_ratings,'director':director,'runtime':runtime,'genre':genres,'cast':cast,'votes':votes,'share':share,'verdict':verdict},index =range(1,len(movie_name)+1) )
df
```

```python
df.to_csv('scraped from imdb.csv',index = False)
```
