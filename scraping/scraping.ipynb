create a data frame using pandas and export it to a csv file

```python
#import essential libraries inn ordre to scrape data
from bs4 import BeautifulSoup
import requests
```

```python
#get the page using the requests package by speifying the url
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_India'

html_content = requests.get(url)
soup = BeautifulSoup(html_content.text,'html')  #parse the content of the webpage with beautiful soup
```

```python
table = soup.find_all('table')[1]
#this is to find content in a html page,as i have wanted to extract information from a table i sprecified the table tag,also i specified a number based on list indexing since i'm intrested to extract information from second table on that page
```

```python
column_titles = table.find_all('th')    #to extract headings from the table
#do not forget to use the variable name as object as using soup again will fetch content from all over the page
```

```python
column_titles_list = [title.text.strip() for title in column_titles]
#run a loop over the extracted headings and store them in a list
```

```python
print(column_titles_list)
```

```python
import pandas as pd
#import pandas to create datframes and  perform data manipulation tasks.
#assign our list of headings as column names to a new dataframe
df = pd.DataFrame(columns=column_titles_list)
df
```

```python
#repeat the same process to dive in deep toextract table rows from table and table data from table rows
column_data = table.find_all('tr')
for row in column_data[1:]:
    row_data =row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    #on every iteration continue to assign the extracted data to the dataframe using the loc function
    length = len(df)
    df.loc[length] = individual_row_data
```
